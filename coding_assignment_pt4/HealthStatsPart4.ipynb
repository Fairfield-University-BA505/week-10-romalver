## W2H Ratio

The waist-hip ratio or waist-to-hip ratio (WHR) is the dimensionless ratio of the circumference of the waist to that of the hips. This is calculated as waist measurement divided by hip measurement (W ÷ H). For example, a person with a 30″ (76 cm) waist and 38″ (97 cm) hips has a waist-hip ratio of about 0.78.
## The Ratio is
$ ratio_{w2h} = \frac{w}{h} $

```python
# Goal: For each row of data calculate and store the w2h_ratio and shape.
import pandas as pd
import numpy as np
import io

df = pd.read_csv('w2h_data.csv')
df
# Step 1: Make sure the columns are in correct data types - refer to previous parts for the correct data types


```

```python
# Step 2: calculate the 'W2H_Ratio' column using the 'Waist' and 'Hip' columns
# Step 3: Create the `Shape` column based on the values of the `W2H_Ratio` column - refer to previous parts for the logic

df['W2H_Ratio'] = df['Waist']/ df['Hip']
df['Shape'] = ['Apple' if d > 1 else "Pear" for d in df["W2H_Ratio"]]
df
```
